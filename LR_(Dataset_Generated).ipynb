## Importing Libraries

```python
import numpy as np
#for numerical computing in Python

import matplotlib.pyplot as plt
#This imports the pyplot module from the Matplotlib library and gives it the alias plt. Matplotlib is a
#popular library for creating static, animated, and interactive visualizations in Python.
```

## Generating Training Dataset

```python
X = 2 * np.random.rand(100,1)
# np.random.rand generates random values (float [0, 1)) from a uniform distribution
```

```python
y = 4 + 3*X + np.random.randn(100,1)
# np.random.randn generates random values from a normal distribution with mean 0 and standard deviation 1
print(y)
```

## Visualizing Training Dataset

```python
plt.plot(X, y, "b.")
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", fontsize=18, rotation = 0)
plt.axis([0, 2, 0, 15])  #giving the range of values on X and Y axis respectively
plt.show()
```

## Adding Bias Term and Finding Best Hyperparameters using Normal Equation

```python
X_b = np.c_[np.ones((100,1 )), X]  # add x0 = 1 to each instance
```

```python
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
```

```python
theta_best
```

```python
X_new = np.array([[0], [2]])
X_new_b = np.c_[np.ones((2, 1)), X_new]
y_predict = X_new_b.dot(theta_best)
y_predict
```

```python
plt.plot(X_new, y_predict, "r-")
plt.plot(X, y, "b.")
plt.axis([0, 2, 0, 15])
plt.show()
```

```python
plt.plot(X_new, y_predict, "r-", linewidth=2, label="Predictions")
plt.plot(X, y, "b.")
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.legend(loc="upper left", fontsize=14)
plt.axis([0, 2, 0, 15])
plt.show()
```

```python
plt.plot(X_new, y_predict, "r-", linewidth=2, label="Predictions")
plt.plot(X, y, "b.")
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.legend(loc="upper left", fontsize=14)
plt.axis([0, 2, 0, 15])
plt.show()
```
